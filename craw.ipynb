```python
import warc
import requests
from contextlib import closing
from bs4 import BeautifulSoup
from langdetect import detect
import csv
```

```python
url = []
l = []
article = []
```

```python
topics = ["sport", "basketball", "hockey", "tennis", "football", "soccer"]
file = warc.open("Lab2/CC-MAIN-20190115225438-20190116011438-00000.warc.wet.gz")
for record in file:
    if record.url is not None: 
        url.append(record.url)
count = 0
for links in url: 
    if any(topic in links for topic in topics):
        try:
            page = requests.get(links)
            soup = BeautifulSoup(page.content, "html.parser")
            texts = ""
            for text in soup.findAll('p'):
                texts += text.getText()
            if any(topic in texts for topic in topics) and detect(texts) == 'en' and texts:
                l.append(links)
                article.append(texts)
                count +=1
        except:
            # print(links)
            continue
```

```python
print(*l, sep='\n')
```

```python
print(count)
print(len(l))
```

```python
with open("commonl.csv", mode = 'w', newline='\n', encoding = "utf-8-sig") as file:
    writer = csv.writer(file, delimiter = ',')
    for link,string in zip(l,article):
        t = string.maketrans("\n\t\r", "   ")
#         string = ' '.join(string.split())
        string = string.translate(t)
        string = string.replace('\\', '')
        writer.writerow([link, string])
file.close()
```
